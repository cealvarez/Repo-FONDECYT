```python
"""
Librerias a utilizar
"""
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
import math
import csv
from scipy import stats
```

```python
"""
Paths
"""

hba1c = "hba1c_3369.csv"

"""
Listas y diccionarios

"""
patients_list = []
cluster_patient = {}
list_transition = []
```

```python


def filter_patients(patients_list):

    """
    Lee el archivo de ruta en chosen_patients y toma la columna 'id_paciente',
    crea una lista con los id's de los pacientes y los agrega en el caso de que no
    esten en la lista patients_list 

    """
    
    data = pd.read_csv(hba1c, sep=';')
    for patient in zip(data['id_paciente']):
        if patient not in patients_list:
            patients_list.append(patient)
            


filter_patients(patients_list)
#Lista con los pacientes que se han hecho el Examen hba1c
patients_list = [x[0] for x in patients_list]
```

```python
"""
Programa

"""

# Leemos ambos archivos
data_hba1c = pd.read_csv(hba1c, sep=';')
data_hba1c['fecha'] = pd.to_datetime(data_hba1c['fecha'], infer_datetime_format=True)

list_hba1c = [list(x) for x in zip(data_hba1c["id_paciente"],data_hba1c["fecha"], data_hba1c["examen_hba1c"] )]
```

```python

def obtener_examenes_pacientes(min_measure, lista_id_pacientes, list_hba1c):
        
    # diccionario que tiene como llave el id del paciente y 
    #como valor los resultados la transicion y los examenes ordenados segun fecha
    resultados = {}
    # recorro todos los pacientes
    for id_paciente in lista_id_pacientes:
        # lista con los examenes del paciente
        filtro_hba1c_id = list(filter(lambda x: x[0] == id_paciente , list_hba1c))
        # si el paciente tiene una sola medicion, no se considera 
        if len(filtro_hba1c_id) > min_measure:
            # recorre cada una de las transiciones 
            resultados[id_paciente] = sorted(filtro_hba1c_id, key = lambda x: x[1])        
        else:
            pass
    return resultados

dic = obtener_examenes_pacientes(1, patients_list, list_hba1c)
```

```python

def hba1c_label(hba1c):
    if hba1c < 7:
        return "Normal"
    elif 7 <= hba1c < 9:
        return "Moderately-High"
    elif 9 <= hba1c < 10.5:
        return "High"
    else:
        return "Very-High"

def hba1c_category(hba1c):
    if hba1c < 7:
        return 1
    elif 7 <= hba1c <= 9:
        return 2
    elif hba1c > 9:
        return 3

def calcular_datos(measure, patients_list, list_hba1c):
        
    patient_curves = {}
    # {id:[Examenes...], id:[Examenes...], ....}
    patient_hba1c = obtener_examenes_pacientes(measure, patients_list, list_hba1c)
    id_pacientes = list(patient_hba1c.keys())
    transiciones = {}
    for patient in patient_hba1c:
            # resultados ordenados por fecha
            resultados = [x[2] for x in patient_hba1c[patient]]
            # fechas de los resultados
            fechas = [x[1] for x in patient_hba1c[patient]]
            
            if len(resultados) > measure:    
                
                diferencia_de_fechas = []
                fecha_inicio = fechas[0]
                for i in range(len(fechas)):
                    diferencia_de_fechas.append(int(str(fechas[i]-fecha_inicio).split(" ")[0]))
                first_hba1c = resultados[0]
                last_hba1c = resultados[-1]
                max_hba1c = max(resultados)
                min_hba1c = min(resultados)
                avg_hba1c = round(np.average(resultados), 2)
                std_hba1c = round(np.std(resultados), 2)
                measurements = len(resultados)
                slopes = []
                deltas_up = []
                deltas_down = []
                days_up_ptge = 0
                transitions = str(hba1c_category(resultados[0]))
                days = diferencia_de_fechas[-1]
                bad_measurements = 0
                good_measurements = 0
                mediate_measurements = 0
                
                # clasificamos resultados de examenes entre buenos, intermedios y malos segun hba1c
                for i in range(len(resultados)):
                    if resultados[i] > 9:
                        bad_measurements += 1
                    elif 7 <= resultados[i] <= 9:
                        mediate_measurements += 1
                    else:
                        good_measurements += 1             
                # calculos los slope
                
                for i in range(len(resultados)-1):

                    if resultados[i+1] < 7 and resultados[i] < 7:
                        delta_y = 0
                    else:
                        delta_y = resultados[i+1] - resultados[i]

                    delta_x = int(str(fechas[i+1] - fechas[i]).split(" ")[0])

                    slope = delta_y/delta_x
                    slopes.append(slope)

                    if delta_y >= 0:
                        deltas_up.append(delta_y)
                        days_up_ptge += delta_x
                    else:
                        deltas_down.append(delta_y)

                    if hba1c_category(resultados[i+1]) != hba1c_category(resultados[i]):
                        transitions += str(hba1c_category(resultados[i+1]))

                max_label = hba1c_label(max_hba1c)
                min_label = hba1c_label(min_hba1c)
                first_label = hba1c_label(first_hba1c)
                last_label = hba1c_label(last_hba1c)
                avg_hba1c_label = hba1c_label(avg_hba1c)
                avg_slope_days = round(np.average(slopes), 5)
                avg_slope_months = round(avg_slope_days*180, 5)
                max_slope = round(max(slopes)*180, 2)
                min_slope = round(min(slopes)*180, 2)
                delta_hba1c_up = round(sum(deltas_up), 2)
                delta_hba1c_down = round(sum(deltas_down), 2)
                delta_hba1c = round(last_hba1c-first_hba1c, 2)
                range_hba1c = round(max_hba1c-min_hba1c, 2)

                days_up_ptge = round((days_up_ptge*100)/int(str(fechas[-1] - fechas[0]).split(" ")[0]), 1)
                if len(deltas_up) == 0:
                    max_delta_up = 0
                else:
                    max_delta_up = round(max(deltas_up), 2)

                slope, intercept, r_value, p_value, std_err = stats.linregress(diferencia_de_fechas, resultados)

                if max_hba1c < 7 or (avg_hba1c < 7 and (mediate_measurements == 1 and bad_measurements == 0)):
                    category = 'Compensado'

                elif slope < 0 and last_hba1c < 7:
                    category = 'Mejora'

                elif max_hba1c <= 9:  
                    category = 'Medianamente Descompensado'

                else:
                    category = 'Altamente Descompensado'

                if not patient in patient_curves:

                    patient_curves[patient] = [patient, first_hba1c, first_label, 
                                                last_hba1c, last_label, max_hba1c, max_label, min_hba1c, 
                                                min_label, avg_slope_days, avg_slope_months, avg_hba1c, 
                                                std_hba1c, avg_hba1c_label, measurements, max_slope, 
                                                min_slope, delta_hba1c_up,delta_hba1c_down, max_delta_up, 
                                                days_up_ptge, delta_hba1c, range_hba1c, category, slope, days]
            else:
                pass
            
    return patient_curves
```

```python

def write_file(min_measurement, delimiter, patients_list, list_hba1c):
    count = 0
    rows = calcular_datos(min_measurement, patients_list, list_hba1c)
    report = open('Camilo.csv', 'w', newline='')
    wr = csv.writer(report, delimiter=delimiter)
    header = ['id_paciente','primera', 'primera_cat', 'ultima', 'ultima_cat', 'max', 'max_cat', 'min', 'min_cat',
              'pendiente_dias_avg', 'pendiente_mes_avg', 'hba1c_promedio', 'hba1c_std', 'hba1c_promedio_cat',
              'mediciones', 'max_pend', 'min_pend', 'aumento_hba1c', 'caida_hba1c', 'max_aumento_hba1c', 'ptje_aumento',
              'delta_hba1c', 'rango', 'cluster', 'pend_reg', 'total_dias']
    wr.writerow(header)
    for patient in rows:
        wr.writerow(rows[patient])
        count += 1
    print('Pacientes reporte: ', count)
    report.close()

    
write_file(1, ";", patients_list, list_hba1c)
```
